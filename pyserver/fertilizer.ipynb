```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Fertilizer Prediction.csv')
df
```

```python
soiltype = {'Sandy': 0, 'Loamy': 1, 'Black': 2, 'Red': 3, 'Clayey': 4}

```

```python
croptype = {
  'Maize': 0,
  'Sugarcane': 1,
  'Cotton': 2,
  'Tobacco': 3,
  'Paddy': 4,
  'Barley': 5,
  'Wheat': 6,
  'Millets': 7,
  'Oil seeds': 8,
  'Pulses': 9,
  'Ground Nuts': 10
}
```

```python
fertilizername = {
  'Urea': 0,
  'DAP': 1,
  '14-35-14': 2,
  '28-28': 3,
  '17-17-17': 4,
  '20-20': 5,
  '10-26-26': 6
}

```

```python
soiltype = {
  'Sandy': 0,
  'Loamy': 1,
  'Black': 2,
  'Red': 3,
  'Clayey': 4
}

```

```python
df['Crop Type'] = df['Crop Type'].map(croptype)
df['Fertilizer Name'] = df['Fertilizer Name'].map(fertilizername)
df['Soil Type'] = df['Soil Type'].map(soiltype)

```

```python
df
```

```python
y = df['Fertilizer Name'].copy()
X = df.drop(['Fertilizer Name'],axis=1).copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,shuffle=True,random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, classification_report

# Assuming X_train, X_test, y_train, y_test are already defined
# Initialize lists to store accuracies
acc = []
acc1 = []
model = []

# Normalize the data
norm = MinMaxScaler().fit(X_train)
X_train_norm = norm.transform(X_train)
X_test_norm = norm.transform(X_test)

# Initialize and train the SVM model
SVM = SVC(kernel='poly', degree=3, C=1)
SVM.fit(X_train_norm, y_train)

# Predict on test data
predicted_values = SVM.predict(X_test_norm)
x = accuracy_score(y_test, predicted_values)
acc.append(x)

# Predict on training data
predicted_values_train = SVM.predict(X_train_norm)
y = accuracy_score(y_train, predicted_values_train)
acc1.append(y)

model.append('SVM')
print("SVM's Accuracy on Test Data is: {:.2f}%".format(x * 100))
print("SVM's Accuracy on Training Data is: {:.2f}%".format(y * 100))

print("Classification Report on Test Data:")
print(classification_report(y_test, predicted_values))
```
